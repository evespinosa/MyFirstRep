### Create the H_Team

From Scraped Player data get all th distinct teams

```python
def hash_row(row):
    # Combine the values from the three columns, converting them to strings and handling NaN values
    combined = str(row['Team'])
    # Generate a hash using SHA-256
    return hashlib.sha256(combined.encode()).hexdigest()

data_base=f'D:/AFL_GRP/DataBase/AFLStats.db'
table_name='ScrapedPlayerStatistics'
connection=sqlite3.connect(database=data_base)

query= f''' 
select Team from  scrapedPlayerStatistics 
group by Team
'''

df_team=pd.read_sql(query,connection)
connection.close()


df_team['Team_Hashkey'] = df_team.apply(hash_row, axis=1)
df_team['Load_Date']=datetime.now().date()
df_team['Record_Source']='ScrapedPlayerStatistics'
df_team=df_team[['Team_Hashkey','Team','Load_Date','Record_Source']]
```

```python
data_base=f'D:/AFL_GRP/DataBase/AFLDataVault.db'
table_name='H_Team'
connection=sqlite3.connect(database=data_base)
df_team.to_sql(table_name, connection, index=False, if_exists='append')
connection.close()
```
