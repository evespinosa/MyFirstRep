From "D:/AFL_GRP/DataBase/AFLStats.db".'ScrapedPlayerStatistics" group game,team statistis and create table.

list of statistics created are ['Disposals','Kicks','Marks','Handballs','Goals','Behinds','Hit Outs','Tackles',Rebounds'
                                ,'Inside 50s','Clearances',Clangers','Frees','Frees Against','Contested Possessions','Uncontested Possessions'
                                ,'Contested Marks','Marks Inside 50','One Percenters','Bounces','Goal Assists']


```python
data_base=f'D:/AFL_GRP/DataBase/AFLStats.db'
connection=sqlite3.connect(data_base)
query=f''' 
with Disposals as
       (select
       year,round,team,sum(value) Disposals
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Disposals'
       group by year,round,team
       ),
Kicks as
       (select
       year,round,team,sum(value) Kicks
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Kicks'
       group by year,round,team
       ),
Marks as
      (select
       year,round,team,sum(value) Marks
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Marks'
       group by year,round,team
       ),
Handballs as
      (select
       year,round,team,sum(value) Handballs
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Handballs'
       group by year,round,team
       ) ,
Goals as
      (select
       year,round,team,sum(value) Goals
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Goals'
       group by year,round,team
       ) , 
Behinds as
      (select
       year,round,team,sum(value) Behinds
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Behinds'
       group by year,round,team
       ) ,
HitOuts as                         
      (select
       year,round,team,sum(value) HitOuts
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Hit Outs'
       group by year,round,team
       ) ,
Tackles as                         
      (select
       year,round,team,sum(value) Tackles
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Tackles'
       group by year,round,team
       ) , 
Rebounds as                         
      (select
       year,round,team,sum(value) Rebounds
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Rebounds'
       group by year,round,team
       ),
Inside50 as                         
      (select
       year,round,team,sum(value) Inside50
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Inside 50s'
       group by year,round,team
       ) ,      
Clearances as                         
      (select
       year,round,team,sum(value) Clearances
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Clearances'
       group by year,round,team
       ),
Clangers as                         
      (select
       year,round,team,sum(value) Clangers
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Clangers'
       group by year,round,team
       ),
Frees as                         
      (select
       year,round,team,sum(value) Frees
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Frees'
       group by year,round,team
       ),
FreesAgainst as                         
      (select
       year,round,team,sum(value) FreesAgainst
       from ScrapedPlayerStatistics 
       where 1=1
       and feature='Frees Against'
       group by year,round,team
       ) ,
ContestedPossessions as                         
      (select
       year,round,team,sum(value) ContestedPossessions
       from ScrapedPlayerStatistics 
       where 1=1
       and feature ='Contested Possessions'
       group by year,round,team
       ),
UncontestedPossessions as                         
      (select
       year,round,team,sum(value) UncontestedPossessions
       from ScrapedPlayerStatistics 
       where 1=1
       and feature ='Uncontested Possessions'
       group by year,round,team
       ),
ContestedMarks as                         
      (select
       year,round,team,sum(value) ContestedMarks
       from ScrapedPlayerStatistics 
       where 1=1
       and feature ='Contested Marks'
       group by year,round,team
       ), 
MarksInside50 as                         
      (select
       year,round,team,sum(value) MarksInside50
       from ScrapedPlayerStatistics 
       where 1=1
       and feature ='Marks Inside 50'
       group by year,round,team
       ),
OnePercenters as                         
      (select
       year,round,team,sum(value) OnePercenters
       from ScrapedPlayerStatistics 
       where 1=1
       and feature ='One Percenters'
       group by year,round,team
       ),
Bounces as                         
      (select
       year,round,team,sum(value) Bounces
       from ScrapedPlayerStatistics 
       where 1=1
       and feature ='Bounces'
       group by year,round,team
       ),
GoalAssists as                         
      (select
       year,round,team,sum(value) GoalAssists
       from ScrapedPlayerStatistics 
       where 1=1
       and feature ='Goal Assists'
       group by year,round,team
       )                                                                                                                                                                                                                                      
select d.*
     ,k.kicks 
     ,m.marks
     ,h.handballs
     ,g.goals
     ,b.behinds
     ,ho.HitOuts
     ,t.tackles
     ,r.rebounds
     ,i50.Inside50
    ,c.clearances
    ,cl.clangers
    ,f.frees
    ,fa.freesagainst
    ,cp.ContestedPossessions
    ,up.UncontestedPossessions
    ,cm.ContestedMarks
    ,mi50.MarksInside50
    ,op.OnePercenters
    ,bo.bounces
    ,ga.GoalAssists
from  Disposals d
join  Kicks k
on    d.year=k.year
and   d.round=k.round
and   d.team=k.team
join Marks M
on    d.year=m.year
and   d.round=m.round
and   d.team=m.team
join Handballs h
on    d.year=h.year
and   d.round=h.round
and   d.team=h.team  
join Goals g
on    d.year=g.year
and   d.round=g.round
and   d.team=g.team 
join behinds b
on    d.year=b.year
and   d.round=b.round
and   d.team=b.team 
join HitOuts ho
on    d.year=ho.year
and   d.round=ho.round
and   d.team=ho.team
join tackles t
on    d.year=t.year
and   d.round=t.round
and   d.team=t.team
join rebounds r
on    d.year=r.year
and   d.round=r.round
and   d.team=r.team
left join inside50 i50
on    d.year=i50.year
and   d.round=i50.round
and   d.team=i50.team
join  clearances c
on    d.year=c.year
and   d.round=c.round
and   d.team=c.team
join  clangers cl
on    d.year=cl.year
and   d.round=cl.round
and   d.team=cl.team
join  frees f
on    d.year=f.year
and   d.round=f.round
and   d.team=f.team
join  freesagainst fa
on    d.year=fa.year
and   d.round=fa.round
and   d.team=fa.team
join  ContestedPossessions cp
on    d.year=cp.year
and   d.round=cp.round
and   d.team=cp.team
join  UncontestedPossessions up
on    d.year=up.year
and   d.round=up.round
and   d.team=up.team
join  contestedMarks cm
on    d.year=cm.year
and   d.round=cm.round
and   d.team=cm.team
join  MarksInside50 mi50
on    d.year=mi50.year
and   d.round=mi50.round
and   d.team=mi50.team
join  OnePercenters op
on    d.year=op.year
and   d.round=op.round
and   d.team=op.team
join  Bounces bo
on    d.year=bo.year
and   d.round=bo.round
and   d.team=bo.team
join  GoalAssists ga
on    d.year=ga.year
and   d.round=ga.round
and   d.team=ga.team
'''

df=pd.read_sql(query,connection)
connection.close()

df
# data_base=f'D:/AFL_GRP/DataBase/AFLDataVault.db'
# table_name='Source_S_Game_Details'
# connection=sqlite3.connect(database=data_base)
# df.to_sql(table_name, connection, index=False, if_exists='append')
# connection.close()
```

```python
def hash_row(row):
    combined = str(row['year']) + str(row['round']) + str(row['team']) 
    # Generate a hash using SHA-256
    return hashlib.sha256(combined.encode()).hexdigest()

def hashdiff_row(row):
    combined = (str(row['Disposals']) + str(row['Kicks']) + str(row['Marks']) + str(row['Handballs']) + str(row['Goals'])+ str(row['Behinds'])
                 + str(row['HitOuts']) + str(row['Tackles']) +str(row['Rebounds']) + str(row['Inside50']) + str(row['Clearances']) +str(row[ 'Clangers']) 
                 + str(row['Frees'])   + str(row['FreesAgainst']) +str(row[ 'ContestedPossessions']) + str(row['UncontestedPossessions'])   
                 + str(row['ContestedMarks']) + str(row['MarksInside50']) + str(row['OnePercenters']) + str(row['Bounces']) + str(row['GoalAssists']))
                            
   
    # Generate a hash using SHA-256
    return hashlib.sha256(combined.encode()).hexdigest()

df['GameTeam_Hashkey'] = df.apply(hash_row, axis=1)
df['Hashdiff']=df.apply(hashdiff_row, axis=1)
df['Record_Source']='ScrapedPlayerStatistics'

df=df[['GameTeam_Hashkey','Disposals','Kicks','Marks','Handballs','Goals','Behinds','HitOuts','Tackles','Rebounds'
                         ,'Inside50','Clearances','Clangers','Frees','FreesAgainst','ContestedPossessions','UncontestedPossessions'
                         ,'ContestedMarks','MarksInside50','OnePercenters','Bounces','GoalAssists','Hashdiff','Record_Source']]
df
```
